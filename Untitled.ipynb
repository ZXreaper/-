### 自定义排序
**将同一区域，按照行递增、列递增的顺序排列起来**
**这样处理之后，我们在删除有退票的情况时：只用从头到尾遍历一遍，即可解决退票的情况，时间复杂度$O(N)$**

```python
import functools

def cmp(a,b):
    if a[1] == b[1]:
        if a[2] == b[2]:
            if a[3] == b[3]:
                if a[0] == b[0]:
                    if a[4] < b[4]:
                        return 1
                    else:
                        return -1
                else:
                    if a[0] < b[0]:
                        return 1
                    else:
                        return -1
            else:
                if a[3] > b[3]:
                    return 1
                else:
                    return -1
        else:
            if a[2] > b[2]:
                return 1
            else:
                return -1
    else:
        if a[1] > b[1]:
            return 1
        else:
            return -1
        
cur_table.sort(key = functools.cmp_to_key(cmp))
```

```python
#-*- coding = utf-8 -*-
#@Time : 2021/5/8 10:03
#@Author : ZXREAPER zhangxu
#@File : main.py
#@Software : PyCharm

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import functools
import math
import seaborn as sns
from dateutil.parser import parse
from matplotlib.colors import LinearSegmentedColormap

# 设置放票时间。该场次什么时间起售
# 起售时间
start_time = parse('2019-9-10/0:0:0')
# 电影上映时间
end_time = parse('2020-01-16/19:30:0')

# 设置场次数量
movie_cnt = 1

'''
各区域编号：
    一楼观众厅:      1
    二楼单号包厢:    2
    二楼双号包厢:    3
    二楼贵宾席V1:    4
    二楼贵宾席V2:    5
    二楼贵宾席V3:    6
    二楼贵宾席V4:    7
    二楼贵宾席V5:    8
    二楼贵宾席V6:    9
    二楼贵宾席V7:    10
    三楼单号包厢:    11
    三楼双号包厢:    12
    三楼观众厅:      13
'''

# 读取数据表并处理
def process_data(file_name):
    org_table = pd.read_excel(file_name, usecols=[0, 12, 15, 16, 17, 19, 20, 21], engine='openpyxl')
    seat_number = {"一楼观众厅": 1, "二楼单号包厢": 2, "二楼双号包厢": 3, "二楼贵宾席V1": 4, "二楼贵宾席V2": 5,
                   "二楼贵宾席V3": 6,"二楼贵宾席V4": 7, "二楼贵宾席V5": 8, "二楼贵宾席V6": 9, "二楼贵宾席V7": 10,
                   "三楼单号包厢": 11, "三楼双号包厢": 12,"三楼观众厅": 13}
    for index, rows in org_table.iterrows():
        org_table['区域名称'][index] = seat_number[rows[5]]
    table = org_table
    table['时间间隔'] = 0
    for index, rows in table.iterrows():
        a = parse(str(table["创建日期"][index]))
        table["时间间隔"][index] = (a - start_time).total_seconds()
    return table


# 读取各区域的座位映射表
def read_distribution_map() -> dict:
    # 加载各区域的映射表
    def read_distribution(file_path) -> list:  # 迷惑行为：不加返回值注释竟然在经过return后list类型变为了NoneType
        a = np.load(file_path, allow_pickle=True)
        res = a.tolist()
        return res

    seat_distribution_map = {}
    for i in range(1, 14):
        file_name = './seat_distribution/NO_' + str(i) + '_seat_distribution.npy'
        temp = read_distribution(file_name)
        seat_distribution_map[i] = temp

    return seat_distribution_map


def cmp(a,b):
    """
    排序方式：
            第一关键字：区域号。递增关系。
            第二关键字：排。递增关系。
            第三关键字：座位号。递增关系。
            第四关键字：数量。退票(-1)在后，买票(1)在前。
            第五关键字：时间间隔。递减关系。

    :param a: first list
    :param b: second list
    :return: 1 or -1
    """
    if a[1] == b[1]:
        if a[2] == b[2]:
            if a[3] == b[3]:
                if a[0] == b[0]:
                    if a[4] < b[4]:
                        return 1
                    else:
                        return -1
                else:
                    if a[0] < b[0]:
                        return 1
                    else:
                        return -1
            else:
                if a[3] > b[3]:
                    return 1
                else:
                    return -1
        else:
            if a[2] > b[2]:
                return 1
            else:
                return -1
    else:
        if a[1] > b[1]:
            return 1
        else:
            return -1

# 解决数据中有退票的情况
def process_refund(table) -> list:
    cur_table = table[['数量', '区域名称', '排', '座', '时间间隔']]
    cur_table = cur_table.values.tolist()
    cur_table.sort(key = functools.cmp_to_key(cmp))

    processed_table = []

    last = cur_table[0]
    cnt = 0
    length = len(cur_table)

    for i in range(1, length):
        if cur_table[i][1] == last[1] and cur_table[i][2] == last[2] and cur_table[i][3] == last[3]:
            if cur_table[i][0] == 1:
                last.append(cur_table[i][4])
            else:
                last.pop()
        else:
            processed_table.append(last)
            last = cur_table[i]
    processed_table.append(last)
    return processed_table

# 获取13张作为分布的numpy矩阵
def get_distribution_numpy_matrix(processed_table) -> list:
    heat_maps = [[]]

    for num in range(1, 14):
        file_name = './seat_distribution/NO_' + str(num) + '_seat_distribution.npy'
        a = np.load(file_name, allow_pickle=True)
        res = a.tolist()
        # 列数
        mx_col = 0
        # 行数
        mx_row = 0
        for i in res:
            for j in i:
                if len(j):
                    mx_col = max(mx_col, j[1])
                    mx_row = max(mx_row, j[0])
        temp = np.zeros([mx_row + 1, mx_col + 1], dtype=int)
        #     exec("heat_map_%s=np.zeros([mx_row+1,mx_col+1],dtype=int)"%num)
        heat_maps.append(temp)

    for item in processed_table:
        region = item[1]
        row_num = item[2]
        seat_num = item[3]
        if len(item) == 5:
            time_len = item[4]
        else:
            time_len = 0
        row = seat[region][row_num][seat_num][0]
        col = seat[region][row_num][seat_num][1]
        # exec("heat_map_%s[row][col] = time_len" % region)
        heat_maps[region][row][col] = time_len

    return heat_maps

# 画热力图
def draw_heat_map(heat_map,processed_table):
    # 获取时间间隔的中位数
    time_slice = []
    for item in processed_table:
        if len(item) == 5:
            time_slice.append(item[4])
    time_slice = np.array(time_slice)

    # 定义颜色条的显示范围
    med = np.median(time_slice)     # median number
    minv = np.min(time_slice)       # min number
    maxv = np.max(time_slice)       # max number

    # 定义渐变颜色条（白色-蓝色-红色）
    my_colormap = LinearSegmentedColormap.from_list("", ["red", "orange", "green"])

    fig = plt.figure()

    # 表1
    ax1 = plt.subplot(5, 1, 1)
    ax1.margins(2, 2)  # Values >0.0 zoom out
    # ax1.set_title('pic-1')
    sns.heatmap(heat_map[1], cbar=False,linewidths=0.05, center=med, vmax=maxv, vmin=minv, mask=heat_map[1] < 10, cmap='rainbow')
    plt.axis('off')
    
    # 表2
    ax2 = plt.subplot(5, 3, 4)
    ax2.margins(2, 2)  # Values >0.0 zoom out
    # ax2.set_title('pic-2')
    sns.heatmap(heat_map[2], cbar=False,linewidths=0.05, center=med, vmax=maxv, vmin=minv, mask=heat_map[2] < 10, cmap='rainbow')
    plt.axis('off')
    
    # 表3
    ax3 = plt.subplot(5, 3, 6)
    ax3.margins(2, 2)  # Values >0.0 zoom out
    # ax3.set_title('pic-3')
    sns.heatmap(heat_map[3], cbar=False,linewidths=0.05, center=med, vmax=maxv, vmin=minv, mask=heat_map[3] < 10, cmap='rainbow')
    plt.axis('off')
    
    # 表4
    ax4 = plt.subplot(5, 7, 18)
    ax4.margins(2, 2)  # Values >0.0 zoom out
    # ax4.set_title('pic-4')
    sns.heatmap(heat_map[4], cbar=False,linewidths=0.05, center=med, vmax=maxv, vmin=minv, mask=heat_map[4] < 10, cmap='rainbow')
    plt.axis('off')
    
    # 表5
    ax5 = plt.subplot(5, 7, 19)
    ax5.margins(2, 2)  # Values >0.0 zoom out
    # ax5.set_title('pic-5')
    sns.heatmap(heat_map[5], cbar=False,linewidths=0.05, center=med, vmax=maxv, vmin=minv, mask=heat_map[5] < 10, cmap='rainbow')
    plt.axis('off')
    
    # 表6
    ax6 = plt.subplot(5, 7, 17)
    ax6.margins(2, 2)  # Values >0.0 zoom out
    # ax6.set_title('pic-6')
    sns.heatmap(heat_map[6], cbar=False,linewidths=0.05, center=med, vmax=maxv, vmin=minv, mask=heat_map[6] < 10, cmap='rainbow')
    plt.axis('off')
    
    # 表7
    ax7 = plt.subplot(5, 7, 20)
    ax7.margins(2, 2)  # Values >0.0 zoom out
    # ax7.set_title('pic-7')
    sns.heatmap(heat_map[7], cbar=False,linewidths=0.05, center=med, vmax=maxv, vmin=minv, mask=heat_map[7] < 10, cmap='rainbow')
    plt.axis('off')
    
    # 表8
    ax8 = plt.subplot(5, 7, 16)
    ax8.margins(2, 2)  # Values >0.0 zoom out
    # ax8.set_title('pic-8')
    sns.heatmap(heat_map[8], cbar=False,linewidths=0.05, center=med, vmax=maxv, vmin=minv, mask=heat_map[8] < 10, cmap='rainbow')
    plt.axis('off')
    
    # 表9
    ax9 = plt.subplot(5, 7, 21)
    ax9.margins(2, 2)  # Values >0.0 zoom out
    # ax9.set_title('pic-9')
    sns.heatmap(heat_map[9], cbar=False,linewidths=0.05, center=med, vmax=maxv, vmin=minv, mask=heat_map[9] < 10, cmap='rainbow')
    plt.axis('off')
    
    # 表10
    ax10 = plt.subplot(5, 7, 15)
    ax10.margins(2, 2)  # Values >0.0 zoom out
    # ax10.set_title('pic-10')
    sns.heatmap(heat_map[10], cbar=False,linewidths=0.05, center=med, vmax=maxv, vmin=minv, mask=heat_map[10] < 10, cmap='rainbow')
    plt.axis('off')
    
    # 表11
    ax11 = plt.subplot(5, 3, 10)
    ax11.margins(2, 2)  # Values >0.0 zoom out
    # ax11.set_title('pic-11')
    sns.heatmap(heat_map[11], cbar=False,linewidths=0.05, center=med, vmax=maxv, vmin=minv, mask=heat_map[11] < 10, cmap='rainbow')
    plt.axis('off')
    
    # 表12
    ax12 = plt.subplot(5, 3, 12)
    ax12.margins(2, 2)  # Values >0.0 zoom out
    # ax12.set_title('pic-12')
    sns.heatmap(heat_map[12], cbar=False,linewidths=0.05, center=med, vmax=maxv, vmin=minv, mask=heat_map[12] < 10, cmap='rainbow')
    plt.axis('off')
    
    # 表13
    ax13 = plt.subplot(5, 1, 5)
    ax13.margins(2, 2)  # Values >0.0 zoom out
    # ax13.set_title('pic-13')
    sns.heatmap(heat_map[13], cbar=False,linewidths=0.05, center=med, vmax=maxv, vmin=minv, mask=heat_map[13] < 10, cmap='rainbow')
    plt.axis('off')
    
    # 显示
    plt.show()
```
